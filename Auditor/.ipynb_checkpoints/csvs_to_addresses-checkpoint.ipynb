```python
import pandas as pd
import numpy as np

import os

import pickle
```

```python

ext = '/data/addresses/'
path = os.getcwd()+ext
```

```python
print(ext+os.listdir(path)[0])
```

```python
df = pd.read_csv("."+ext+os.listdir(path)[0])

fails = []

for file_name in os.listdir(path)[1:]:
    try:
        new_info = pd.read_csv("."+ext+file_name)
        df = df.append(new_info, ignore_index = True)
    except:
        fails.append(file_name)
```

```python
fails
```

```python
df
```

```python
pickle_out = open(os.getcwd()+'/data/pickle_dumps/addresses.pickle', 'wb')
pickle.dump(df,pickle_out)
pickle_out.close()
```

```python
pickle_in = open(os.getcwd()+'/data/pickle_dumps/addresses.pickle', 'rb')
recovered_df = pickle.load(pickle_in)
pickle_in.close()
```
